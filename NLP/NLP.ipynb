```python
corpus = []
for i in range(0,len(df)):
    review = re.sub('[^a-zA-Z0-9]'," ", df['message'][i] )
    review =review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if word not in stop_word_list]
    review = " ".join(review)
    corpus.append(review)
```
